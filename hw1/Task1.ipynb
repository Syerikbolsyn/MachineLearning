```python
import numpy as np
import pandas as pd
import os
```

```python
def cos_distance(a,b):
    return 1 - a@b/(np.sqrt(a@a)*np.sqrt(b@b))
```

```python
path = 'sentences.txt'
f = open(path, 'r')

lines = f.readlines()
lines
```

```python
import re
tokens = []

for line in lines:

    a = [string for string in re.split('[^a-z]',line.lower()) if string != '']
    tokens.append(a)
tokens
```

```python
unique_words = list(set([word for lines in tokens for word in lines]))
word_idx = {word : i for i, word in enumerate(unique_words)}
word_idx
```

```python
count_matrix = np.zeros((len(lines), len(unique_words)))
for i,line in enumerate(tokens):
    for word in line:
        count_matrix[i,word_idx[word]] = line.count(word)
count_matrix
```

```python
distances = np.array([cos_distance(count_matrix[0], count_matrix[i]) for i in range(1, count_matrix.shape[0])])
distances
```

```python
nearest_sentences = np.argsort(distances)[:2] + 1
nearest_sentences
```
